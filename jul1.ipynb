```python
#Q1. Check Quadrant of a Point (x, y)
x = float(input("Enter X coordinate: "))
y = float(input("Enter Y coordinate: "))

if x > 0 and y > 0:
    print("First Quadrant")
elif x < 0 and y > 0:
    print("Second Quadrant")
elif x < 0 and y < 0:
    print("Third Quadrant")
elif x > 0 and y < 0:
    print("Fourth Quadrant")
elif x == 0 and y != 0:
    print("On Y-Axis")
elif y == 0 and x != 0:
    print("On X-Axis")
else:
    print("At the Origin")
```

```python
#Q2. Check if Three Sides Form a Valid Triangle
a = float(input("Enter side 1: "))
b = float(input("Enter side 2: "))
c = float(input("Enter side 3: "))

if (a + b > c) and (a + c > b) and (b + c > a):
    print("Valid Triangle")
else:
    print("Invalid Triangle")
```

```python
#Q4. Check Student Grade Based on Marks
marks = float(input("Enter student marks (0-100): "))

if marks < 0 or marks > 100:
    print("Invalid marks entered.")
elif marks >= 85:
    print("Grade: A")
elif marks >= 70:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```

```python
#Q5. Calculate Electricity Bill with Slab Rates
units = float(input("Enter electricity units consumed: "))
bill = 0.0

if units <= 100:
    bill = 0
elif units <= 300:
    bill = (units - 100) * 5
else:
    bill = (200 * 5) + ((units - 300) * 10)

# Apply a 10% surcharge if consumption exceeds 500 units
if units > 500:
    bill += bill * 0.10

print(f"Total Electricity Bill: ₹{bill:.2f}")
```

```python
#Q6. Product of Digits of a Given Number
num = int(input("Enter a number: "))
temp = abs(num)
product = 1 if temp > 0 else 0

while temp > 0:
    digit = temp % 10
    product *= digit
    temp //= 10

print(f"Product of digits: {product}")
```

```python
#Q7. Sum of All Factors of a Given Number
num = int(input("Enter a number: "))
factor_sum = 0

for i in range(1, abs(num) + 1):
    if num % i == 0:
        factor_sum += i

print(f"Sum of factors: {factor_sum}")
```

```python
#Q8. Convert Decimal to Binary
decimal_num = int(input("Enter a decimal number: "))

# Built-in way: bin(decimal_num)[2:]
# Manual algorithm:
if decimal_num == 0:
    binary_str = "0"
else:
    binary_str = ""
    temp = abs(decimal_num)
    while temp > 0:
        binary_str = str(temp % 2) + binary_str
        temp //= 2
    if decimal_num < 0:
        binary_str = "-" + binary_str

print(f"Binary representation: {binary_str}")
```

```python
#Q9. Find the Largest Digit in a Given Number
num = int(input("Enter a number: "))
temp = abs(num)
largest_digit = 0

while temp > 0:
    digit = temp % 10
    if digit > largest_digit:
        largest_digit = digit
    temp //= 10

print(f"Largest digit: {largest_digit}")
```

```python
#Q10. Print All Prime Numbers from 1 to N
n = int(input("Enter the value of N: "))

print(f"Prime numbers between 1 and {n}:")
for num in range(2, n + 1):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
print()
```
